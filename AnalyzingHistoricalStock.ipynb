```python
import yfinance as yf

# Extracting stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Extracting financials data which includes revenue
tesla_financials = tesla.financials
print(tesla_financials.loc['Total Revenue'])

```

```python
import yfinance as yf

# Extracting stock data
gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="max")

# Extracting financials data which includes revenue
gamestop_financials = gamestop.financials
print(gamestop_financials.loc['Total Revenue'])

```

```python
import matplotlib.pyplot as plt

# Plotting Tesla stock price and revenue
fig, ax1 = plt.subplots()

ax1.plot(tesla_data.index, tesla_data['Close'], 'b-', label='Tesla Stock Price')
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(tesla_financials.columns, tesla_financials.loc['Total Revenue'], 'r-', label='Tesla Revenue')
ax2.set_ylabel('Revenue', color='r')
ax2.tick_params('y', colors='r')

fig.tight_layout()
plt.title('Tesla Stock Price and Revenue')
plt.show()

```

```python
import matplotlib.pyplot as plt

# Plotting GameStop stock price and revenue
fig, ax1 = plt.subplots()

ax1.plot(gamestop_data.index, gamestop_data['Close'], 'b-', label='GameStop Stock Price')
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(gamestop_financials.columns, gamestop_financials.loc['Total Revenue'], 'r-', label='GameStop Revenue')
ax2.set_ylabel('Revenue', color='r')
ax2.tick_params('y', colors='r')

fig.tight_layout()
plt.title('GameStop Stock Price and Revenue')
plt.show()
```
